# OpenCV Contours

 - cv2.findContours(image, mode, method) : 이미지에서 Contour들을 찾는 함수
  - mode : Countours들을 찾는 방법

## 1) RETR_EXTERNAL : 바깥쪽 Line만 찾기
## 2) RETR_LIST : 모든 Line을 찾지만. Hierachy 구성 x
## 3) RETR_TREE : 모든 Line을 찾으며, 모든 Hierarchy 구성 O
 - method : Countours를 찾는 근사치 방법

## 1) CHAIN_APPROX_NONE: 모든 COUNTOURS 포인트 저장
## 2) CHAIN_APPROX_SIMPLE : Countours Line을 그릴 수 있는 포인트 저장
## 3) 2) CHAIN_APPROX_SIMPLE: Contour Line을 그릴 수 있는 포인트만 저장
 - 입력 이미지는 Gray Scale Thresholod 전처리 과정이 필요합니다.

-cv2.drawContours(image, contours, contoru_index, color, thickness) : Countorus를 그리스는 함수
 - contours:index : 그리고자 하는 함수 Contorus Line (전체 -1)

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('gray.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(image_gray, 127, 255, 0)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[1]
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 4)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

```
