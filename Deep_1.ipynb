# 데이터 시각화

## 데이터 과학자가 갖춰야할 기본 기술 중 하나는 데이터 시각화이다.
 - 데이터 탐색(exploration)
 - 데이터 전달(communication)


### matplotlib

 - 웹을 위한 복잡하고 인터랙티브한 시각화를 만들고 싶다면 가장 좋은 선택은 아닐 수 있지만, 막대 그래프, 선 그래프, 또는 산점도를 그릴 떄는 나쁘지 않다.


 

```
import matplotlib.pyplot as plt

years= [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.2, 1075.9, 2862.1, 5979.6, 10289.7, 14958.3]

# x축에 연도, y축에 GDP가 있는 선 그래프 만듬.
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

#제목
plt.title('Nominal GDP')

#Y축에 레이블 추가
plt.ylabel('Billions of $')

plt.show()
```

```
# 막대
import matplotlib.pyplot as plt

movies = ['Annile Hall', 'Ben-Hur', 'Casablanca','West Size Story']  # x축.
num_oscars = [5, 11, 3, 8, 10] # y축

# 막대 너비의 기본값이 0.8이므로
# 막대가 가운데로 올 수 있도록 왼쪽 좌표에 0.1씩 더 해주자.

xs = [i + 0.1 for i, _ in enumerate(movies)]

# 왼편으로부터 x축의 위치가 xs이고 높이가 num_oscars인 막대.
plt.bar(xs,num_oscars)
plt.ylabel('# of Academy Awards')
plt.title('My Favorite Movies')

# 막대의 가운데에 오도록 영화 제목 레이블.
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```
import matplotlib.pyplot as plt

mentions = [500, 505]
years = [2013, 2014]

plt.bar([2012.6,2013.6], mentions, 0.8)
plt.xticks(years)
plt.ylabel("# Of times I heard Someone say 'data science")

plt.ticklabel_format(useOffset=False)

plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```
